```python
import pandas as pd
```

```python
df_datapercbr_bySite = pd.read_csv("data/datapercbr_bySite.tsv", sep="\t", header=0)
df_datapercbr_bySite.head()
```

```python
df_datapercbr_bySite_fill_by_zero = df_datapercbr_bySite.fillna(0)
df_datapercbr_bySite_fill_by_zero.head()
```

```python
exclude_conditions = ["DMEM...R1", "DMEM...R2"] #"EV.G...R1", "EV.G...R2", "V.d...R1", "V.d...R2", "V.r...R1", "V.r...R2"]
df_exclude_conditions = df_datapercbr_bySite_fill_by_zero.drop(columns=exclude_conditions)
df_exclude_conditions.head()
```

```python
transpose_df = df_exclude_conditions.transpose()
transpose_df
```

```python
condition_names = transpose_df.index.tolist()
condition_names = condition_names[1:]
condition_names
```

```python
class_names = list()
for item in condition_names:
    if item.startswith("EV."):
        class_names.append("EV") 
    if item.startswith("P."):
        class_names.append("Control")
    if item.startswith("T8."):
        class_names.append("T8")
class_names
```

```python
feature_names = transpose_df.iloc[0, :]
feature_names
```

```python
transpose_df_data = transpose_df[1:]
transpose_df_data
```

```python
transpose_df_data["Class"] = class_names
transpose_df_data
```

```python
transpose_df_data_no_index = transpose_df_data.reset_index(drop=True)
transpose_df_data_no_index
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# Fit and transform
type_labels = le.fit_transform(transpose_df_data_no_index["Class"])
type_labels
```

```python
# Print results
print(type_labels)
print(le.classes_)  # Check class order
```

```python
transpose_df_data_no_index["Class"] = type_labels
transpose_df_data_no_index
```

```python
import numpy as np
import xgboost as xgb
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score

def stratified_kfold_xgboost(X, y, n_splits=5, params=None, num_boost_round=100, early_stopping_rounds=10):
    
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    scores = []
    models = []

    for fold, (train_idx, valid_idx) in enumerate(skf.split(X, y)):
        print(f"Training Fold {fold+1}/{n_splits}...")

        # Split data
        X_train, X_valid = X[train_idx], X[valid_idx]
        y_train, y_valid = y[train_idx], y[valid_idx]
        
        print("X shape:", X_train.shape)
        print("y shape:", y_train.shape)

        # Set default parameters if none are provided
        if params is None:
            params = {
                'objective': 'multi:softmax',  # 'binary:logistic' for binary classification
                'num_class': len(np.unique(y)),  # Needed for multi-class classification
                'eval_metric': 'mlogloss',  # 'logloss' for binary classification
                'learning_rate': 0.1,
                'max_depth': 6,
                'subsample': 0.8,
                'colsample_bytree': 0.8,
                'random_state': 42
            }

        # Initialize and train XGBoost model
        model = xgb.XGBClassifier(**params, n_estimators=num_boost_round, use_label_encoder=False)
        model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)], 
                  early_stopping_rounds=early_stopping_rounds, verbose=False)

        # Make predictions
        y_pred = model.predict(X_valid)

        # Calculate accuracy
        acc = accuracy_score(y_valid, y_pred)
        scores.append(acc)
        #models.append(model)

        print(f"Fold {fold+1} Accuracy: {acc:.4f}")

    print(f"Mean Accuracy: {np.mean(scores):.4f}, Std: {np.std(scores):.4f}")
    return scores, model
```
