```python
import numpy as np
import xgboost as xgb
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score

def stratified_kfold_xgboost(X, y, n_splits=5, params=None, num_boost_round=100, early_stopping_rounds=10):
    """
    Perform stratified k-fold cross-validation with XGBoost (without using DMatrix).

    Parameters:
    - X: np.array or pd.DataFrame, feature matrix
    - y: np.array or pd.Series, target labels
    - n_splits: int, number of folds (default: 5)
    - params: dict, XGBoost hyperparameters (default: None)
    - num_boost_round: int, number of boosting rounds (default: 100)
    - early_stopping_rounds: int, rounds without improvement before stopping (default: 10)

    Returns:
    - scores: list, accuracy scores for each fold
    - models: list, trained XGBoost models for each fold
    """
    
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    scores = []
    models = []

    for fold, (train_idx, valid_idx) in enumerate(skf.split(X, y)):
        print(f"Training Fold {fold+1}/{n_splits}...")

        # Split data
        X_train, X_valid = X[train_idx], X[valid_idx]
        y_train, y_valid = y[train_idx], y[valid_idx]
        
        print("X shape:", X_train.shape)
        print("y shape:", y_train.shape)

        # Set default parameters if none are provided
        if params is None:
            params = {
                'objective': 'multi:softmax',  # 'binary:logistic' for binary classification
                'num_class': len(np.unique(y)),  # Needed for multi-class classification
                'eval_metric': 'mlogloss',  # 'logloss' for binary classification
                'learning_rate': 0.1,
                'max_depth': 6,
                'subsample': 0.8,
                'colsample_bytree': 0.8,
                'random_state': 42
            }

        # Initialize and train XGBoost model
        model = xgb.XGBClassifier(**params, n_estimators=num_boost_round, use_label_encoder=False)
        model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)], 
                  early_stopping_rounds=early_stopping_rounds, verbose=False)

        # Make predictions
        y_pred = model.predict(X_valid)

        # Calculate accuracy
        acc = accuracy_score(y_valid, y_pred)
        scores.append(acc)
        #models.append(model)

        print(f"Fold {fold+1} Accuracy: {acc:.4f}")

    print(f"Mean Accuracy: {np.mean(scores):.4f}, Std: {np.std(scores):.4f}")
    return scores, model
```
