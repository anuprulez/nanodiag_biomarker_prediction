```python
import pandas as pd
```

```python
df_datapercbr_bySite = pd.read_csv("data/datapercbr_bySite.tsv", sep="\t", header=0)
df_datapercbr_bySite.head()
```

```python
df_datapercbr_bySite_fill_by_zero = df_datapercbr_bySite.fillna(0)
df_datapercbr_bySite_fill_by_zero.head()
```

```python
exclude_conditions = ["DMEM...R1", "DMEM...R2"] #"EV.G...R1", "EV.G...R2", "V.d...R1", "V.d...R2", "V.r...R1", "V.r...R2"]
df_exclude_conditions = df_datapercbr_bySite_fill_by_zero.drop(columns=exclude_conditions)
df_exclude_conditions.head()
```

```python
transpose_df = df_exclude_conditions #.transpose()
transpose_df
```

```python
modification_names = transpose_df.iloc[:, 0]
modification_names
```

```python
cols_control_celllines = ["P.d...R1", "P.d...R2", "P.r...R1", "P.r...R2", "T8.d...R1", "T8.d...R2", "T8.r...R1", "T8.r...R2", "V.d...R1",	"V.d...R2", "V.r...R1", "V.r...R2"]
# corresponding to column names P, Q, R, S, T, U, AF, AG, AJ, AK, AL, AM

cols_empty_vectors_EVS = ["EV.dep...R1", "EV.dep...R2",	"EV.F1_2_3...R1", "EV.F1_2_3...R2", "EV.F4_5...R1", "EV.F4_5...R2", "EV.G...R1", "EV.G...R2", "EV.L...R1", "EV.L...R2", "EV.S...R1", "EV.S...R2"]

cols_t8s = ["T8.dep...R1", "T8.dep...R2", "T8.F1_2_3...R1", "T8.F1_2_3...R2",	"T8.F4_5...R1",	"T8.F4_5...R2",	"T8.G...R1", "T8.G...R2", "T8.L...R1",	"T8.L...R2", "T8.S...R1", "T8.S...R2"]
```

```python
all_cols = cols_control_celllines + cols_empty_vectors_EVS
all_cols = all_cols + cols_t8s
print(len(all_cols)), set(cols_control_celllines).intersection(set(cols_empty_vectors_EVS)), set(cols_control_celllines).intersection(set(cols_t8s)), set(cols_empty_vectors_EVS).intersection(set(cols_t8s))
```

```python
import numpy as np
df_control = transpose_df[cols_control_celllines]
df_control
```

```python
med_control = df_control.median(axis=1)
mean_control = df_control.mean(axis=1)

df_control["median"] = med_control.tolist()
df_control["mean"] = mean_control.tolist()
df_control
```

```python
df_empty_vectors = transpose_df[cols_empty_vectors_EVS]
df_empty_vectors
```

```python
med_control = df_empty_vectors.median(axis=1)
mean_control = df_empty_vectors.mean(axis=1)

df_empty_vectors["median"] = med_control.tolist()
df_empty_vectors["mean"] = mean_control.tolist()
df_empty_vectors
```

```python
df_t8s = transpose_df[cols_t8s]
df_t8s
```

```python
med_control = df_t8s.median(axis=1)
mean_control = df_t8s.mean(axis=1)

df_t8s["median"] = med_control.tolist()
df_t8s["mean"] = mean_control.tolist()
df_t8s
```

```python
agg_condition = "mean"
df_conditions = pd.concat([df_control[agg_condition], df_empty_vectors[agg_condition], df_t8s[agg_condition]], axis=1)
df_conditions.columns = ["control", "empty_vectors_EV", "T8s"]
df_conditions
```

```python
df_conditions["modification_names"] = modification_names
df_conditions
```

```python
import numpy as np

def compute_log2fc(df, condition_A, condition_B, pseudocount=0.01):
    # Convert percentages to proportions and add pseudocount
    prop_A = (df[condition_A] / 100) + pseudocount
    prop_B = (df[condition_B] / 100) + pseudocount
    
    # Compute log2 fold change
    log2fc = np.log2(prop_B / prop_A)
    
    return log2fc

# Compute log2 fold change
df_conditions["log2FC_EVs/Ctrl"] = compute_log2fc(df_conditions, "control", "empty_vectors_EV")
df_conditions["log2FC_T8s/Ctrl"] = compute_log2fc(df_conditions, "control", "T8s")
df_conditions["log2FC_T8s/EVs"] = compute_log2fc(df_conditions, "empty_vectors_EV", "T8s")
```

```python
df_conditions
```

```python
df_conditions.to_csv("log2foldchange_histone_modifications.csv", sep="\t", index=None)
```

```python
diff_quantified_t8_VS_empty_vector = df_conditions[ (df_conditions["log2FC_T8s/EVs"] < -1) | (df_conditions["log2FC_T8s/EVs"] >= 1) ]
diff_quantified_t8_VS_empty_vector = diff_quantified_t8_VS_empty_vector.sort_values(by="log2FC_T8s/EVs", ascending=False)
diff_quantified_t8_VS_empty_vector.to_csv("diff_quantified_t8_VS_empty_vectors.csv", sep="\t", index=None)
diff_quantified_t8_VS_empty_vector
```

```python
diff_quantified_t8_VS_controls = df_conditions[ (df_conditions["log2FC_T8s/Ctrl"] < -1) | (df_conditions["log2FC_T8s/Ctrl"] >= 1) ]
diff_quantified_t8_VS_controls = diff_quantified_t8_VS_controls.sort_values(by="log2FC_T8s/Ctrl", ascending=False)
diff_quantified_t8_VS_controls.to_csv("diff_quantified_t8_VS_control.csv", sep="\t", index=None)
diff_quantified_t8_VS_controls
```

```python
diff_quantified_empty_vector_VS_control = df_conditions[ (df_conditions["log2FC_EVs/Ctrl"] < -1) | (df_conditions["log2FC_EVs/Ctrl"] >= 1) ]
diff_quantified_empty_vector_VS_control= diff_quantified_empty_vector_VS_control.sort_values(by="log2FC_T8s/Ctrl", ascending=False)
diff_quantified_empty_vector_VS_control.to_csv("diff_quantified_empty_vectors_VS_control.csv", sep="\t", index=None)
diff_quantified_empty_vector_VS_control
```
