```python
import numpy as np
from bioinfokit import analys, visuz
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def compute_log2fc(df, condition_A, condition_B, pseudocount=0.01):
    # Convert percentages to proportions and add pseudocount
    prop_A = (df[condition_A] / 100) + pseudocount
    prop_B = (df[condition_B] / 100) + pseudocount
    
    # Compute log2 fold change
    log2fc = np.log2(prop_A /prop_B)
    
    return log2fc

def plot_volcano(df, name):

    # Thresholds
    pval_threshold = 0.05
    logfc_threshold = 1

    df['color'] = 'gray'
    p_value_field = 'p_value'
    df.loc[(df[p_value_field] < pval_threshold) & (df['log2FC'] > logfc_threshold), 'color'] = 'red'    # upregulated
    df.loc[(df[p_value_field] < pval_threshold)& (df['log2FC'] < -logfc_threshold), 'color'] = 'blue'   # downregulated

    # Plot
    plt.figure(figsize=(8, 6))
    plt.scatter(df['log2FC'], df[p_value_field], c=df['color'], label=df["Modification_names"], alpha=0.8, edgecolors='k')

    for i, row in df.iterrows():
        if row["color"] in ["red", "blue"]:
            plt.annotate(row["Modification_names"], (row["log2FC"], row[p_value_field]), fontsize=9, alpha=0.8, rotation=90)

    # Add significance threshold lines
    plt.axhline(pval_threshold, color='black', linestyle='--', linewidth=1)
    plt.axvline(logfc_threshold, color='black', linestyle='--', linewidth=1)
    plt.axvline(-logfc_threshold, color='black', linestyle='--', linewidth=1)

    plt.xlabel('Log2 Fold Change')
    plt.ylabel('P-value (<0.05 significance threshold)')
    plt.title(f'Log2FC vs P-value: {name.replace("HT1080-", "")}')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(output_path + "volcano/Plot_for_{}.png".format(name.replace("HT1080-", "")))
    plt.show()


def plot_umap(df, name):
    print(df.head())
```

```python
from scipy.stats import mannwhitneyu, false_discovery_control, ttest_ind, ttest_ind_from_stats
import seaborn as sns
from statsmodels.stats.multitest import multipletests

def compute_pvalues(df, group1_cols, group2_cols):

    print(group1_cols, group2_cols)
    p_values = []
    for i, row in df.iterrows():
        g1 = row[group1_cols].values.astype(float)
        g2 = row[group2_cols].values.astype(float)
        t_stat, p = ttest_ind(g1, g2, equal_var=False)
        p_values.append(p)

    print(f"T-statistic: {t_stat}, P-value: {p_values}")

    return p_values


def compute_differential_modifications():
    agg_condition = "mean"
    file_name = output_path + "differential_histone_modification.xlsx"
    
    first_condition =  [Tspan8_lEV, Tspan8_sEV, Tspan8_EVdep, Tspan8_F1, Tspan8_F3, V_lEV, V_sEV, V_EVdep, V_F1, V_F3, Tspan8_F1, Tspan8_F3, F1, F3, V_F1, V_F3] 
    second_condition = [V_lEV, V_sEV, V_EVdep, V_F1, V_F3, lEV, sEV, EVdep, F1, F3, Tspan8_sEV, Tspan8_EVdep, sEV, EVdep, V_sEV, V_EVdep]

    writer = pd.ExcelWriter(file_name, engine="xlsxwriter")
    for idx, v_item in enumerate(second_condition):
        v_name = v_item[0]
        df_V_condition = transpose_df[v_item]
        df_V_condition = df_V_condition.replace(0, np.nan)
        df_V_condition = df_V_condition[v_item].replace({',': ''}, regex=True).astype(float)
        mean_v = df_V_condition.mean(axis=1)
        df_V_condition["mean"] = mean_v.tolist()
        
        t8_item =  first_condition[idx]
        t8_name = t8_item[0]

        p_values = compute_pvalues(transpose_df, t8_item, v_item)
        transpose_df["p_value"] = p_values
        #adj_pval = multipletests(transpose_df['p_value'].fillna(1), method='fdr_bh')[1]
        #transpose_df["adj_pval"] = adj_pval

        df_t8_condition = transpose_df[t8_item]
        df_t8_condition = df_t8_condition.replace(0, np.nan)
        df_t8_condition = df_t8_condition[t8_item].replace({',': ''}, regex=True).astype(float)
        
        mean_t8 = df_t8_condition.mean(axis=1)
        df_t8_condition["mean"] = mean_t8.tolist()

        df_T8_V = pd.concat([df_t8_condition[agg_condition], df_V_condition[agg_condition]], axis=1)

        df_T8_V.columns = [t8_name, v_name]

        df_T8_V["modification_names"] = modification_names
        df_T8_V["p_value"] = transpose_df["p_value"]
        #df_T8_V["adj_pval"] = transpose_df["adj_pval"]

        df_T8_V = df_T8_V[(df_T8_V[t8_name] > 0) & (df_T8_V[v_name] > 0)]
 
        # Compute log2FC
        log2_name = "log2FC_{}/{}".format(t8_name, v_name)
        df_T8_V[log2_name] = compute_log2fc(df_T8_V, t8_name, v_name)

        # Filter out differential ones
        df_T8_V = df_T8_V[(df_T8_V[log2_name] < -1) | (df_T8_V[log2_name] >= 1)]

        df_T8_V["adj_pval"] = multipletests(df_T8_V['p_value'], method='fdr_bh')[1]

        print(df_T8_V)

        adj_pval = false_discovery_control(df_T8_V["p_value"], method='by')
        df_T8_V["adj_pval"] = adj_pval

        df_T8_V = df_T8_V.sort_values(by=log2_name, ascending=False)

        df_T8_V_vis = df_T8_V.copy()

        df_T8_V_vis.columns = ["value1", "value2", "Modification_names", "p_value", "log2FC", "adj_pval"]

        print(df_T8_V_vis.head())
        
        tab_name = "{}_VS_{}".format(t8_name, v_name)

        plot_volcano(df_T8_V_vis, tab_name)

        plot_umap(df_T8_V_vis, tab_name)

        df_T8_V.to_csv(output_path + "log2FC_histone_modifications_{}.csv".format(tab_name), sep="\t", index=None)

        e_tabname = tab_name.replace("HT1080-", "")

        df_T8_V.to_excel(writer, sheet_name=e_tabname, index=False)

        #if idx == 1:
        #    break
        #break
    writer.close()
compute_differential_modifications()
```
