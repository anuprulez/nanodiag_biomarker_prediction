```python
import pandas as pd
```

```python
df_datapercbr_bySite = pd.read_csv("data/datapercbr_bySite.tsv", sep="\t", header=0)
df_datapercbr_bySite.head()
```

```python
df_datapercbr_bySite_fill_by_zero = df_datapercbr_bySite.fillna(0)
df_datapercbr_bySite_fill_by_zero.head()
```

```python
exclude_conditions = ["DMEM...R1", "DMEM...R2", "EV.G...R1", "EV.G...R2", "V.d...R1", "V.d...R2", "V.r...R1", "V.r...R2"]
df_exclude_conditions = df_datapercbr_bySite_fill_by_zero.drop(columns=exclude_conditions)
df_exclude_conditions.head()
```

```python
transpose_df = df_exclude_conditions.transpose()
transpose_df
```

```python
condition_names = transpose_df.index.tolist()
condition_names = condition_names[1:]
condition_names
```

```python
class_names = list()
for item in condition_names:
    if item.startswith("EV.") or item.startswith("P."):
        class_names.append("Control_EV")
    if item.startswith("T8."):
        class_names.append("T8")
class_names
```

```python
feature_names = transpose_df.iloc[0, :]
feature_names
```

```python
transpose_df_data = transpose_df[1:]
transpose_df_data
```

```python
transpose_df_data["Class"] = class_names
transpose_df_data
```

```python
transpose_df_data_no_index = transpose_df_data.reset_index(drop=True)
transpose_df_data_no_index
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()

# Fit and transform
binary_labels = lb.fit_transform(transpose_df_data_no_index["Class"])
```

```python
transpose_df_data_no_index["Class"] = binary_labels
transpose_df_data_no_index
```

```python

df_labels = transpose_df_data_no_index["Class"]
df_features = transpose_df_data_no_index.drop(columns=["Class"])
df_features = df_features.astype(float)
df_features
```

```python
X = df_features.to_numpy()
y = df_labels.to_numpy()
```

```python
print(f"X shape: {X.shape}, y shape: {y.shape}")
```

```python
import umap
import seaborn as sns
import matplotlib.pyplot as plt


n_neighbors=35 #10 #10 #5
min_dist=0.999 #0.99 #0.3
metric='correlation'

#embeddings = umap.UMAP(n_neighbors=n_neighbors, min_dist=min_dist, metric='correlation').fit_transform(nebit_features)
embeddings = umap.UMAP().fit_transform(X)
#n_neighbors=n_neighbors, 
#min_dist=min_dist, 
#metric='correlation

# Create a DataFrame with UMAP components and labels
data = {"UMAP1": embeddings[:, 0], "UMAP2": embeddings[:, 1], "Label": y}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.scatterplot(x="UMAP1", y="UMAP2", hue="Label", data=df, palette="viridis", s=50, alpha=1.0)
plt.title("UMAP Visualization")
plt.grid(True)
#plt.savefig("../data/output/umap_NeDBIT_features_{}_{}_all.pdf".format(n_neighbors, min_dist))

plt.show()
```
