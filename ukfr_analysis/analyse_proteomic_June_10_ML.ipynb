```python
import numpy as np
from bioinfokit import analys, visuz
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def compute_log2fc(df, condition_A, condition_B, pseudocount=0.01):
    # Convert percentages to proportions and add pseudocount
    prop_A = (df[condition_A] / 100) + pseudocount
    prop_B = (df[condition_B] / 100) + pseudocount
    
    # Compute log2 fold change
    log2fc = np.log2(prop_A /prop_B)
    
    return log2fc

def plot_volcano(df, name):

    # Thresholds
    pval_threshold = 0.05
    logfc_threshold = 1

    df['color'] = 'gray'
    p_value_field = 'p_value'
    df.loc[(df[p_value_field] < pval_threshold) & (df['log2FC'] > logfc_threshold), 'color'] = 'red'    # upregulated
    df.loc[(df[p_value_field] < pval_threshold)& (df['log2FC'] < -logfc_threshold), 'color'] = 'blue'   # downregulated

    # Plot
    plt.figure(figsize=(8, 6))
    plt.scatter(df['log2FC'], df[p_value_field], c=df['color'], label=df["Modification_names"], alpha=0.8, edgecolors='k')

    for i, row in df.iterrows():
        if row["color"] in ["red", "blue"]:
            plt.annotate(row["Modification_names"], (row["log2FC"], row[p_value_field]), fontsize=9, alpha=0.8, rotation=90)

    # Add significance threshold lines
    plt.axhline(pval_threshold, color='black', linestyle='--', linewidth=1)
    plt.axvline(logfc_threshold, color='black', linestyle='--', linewidth=1)
    plt.axvline(-logfc_threshold, color='black', linestyle='--', linewidth=1)

    plt.xlabel('Log2 Fold Change')
    plt.ylabel('P-value (<0.05 significance threshold)')
    plt.title(f'Log2FC vs P-value: {name.replace("HT1080-", "")}')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(output_path + "volcano/Plot_for_{}.png".format(name.replace("HT1080-", "")))
    plt.show()


def plot_umap(df, name):
    print(df.head())
```

```python
from scipy.stats import mannwhitneyu, false_discovery_control, ttest_ind, ttest_ind_from_stats
import seaborn as sns
from statsmodels.stats.multitest import multipletests
from scipy.stats import ttest_1samp

def compute_pvalues(df, group1_cols, group2_cols):

    print(group1_cols, group2_cols)
    p_values = []
    for i, row in df.iterrows():
        g1 = row[group1_cols].values.astype(float)
        g2 = row[group2_cols].values.astype(float)
        t_stat, p = ttest_ind(g1, g2, equal_var=False)
        p_values.append(p)

    print(f"T-statistic: {t_stat}, P-value: {p_values}")

    return p_values


def compute_pvalues_one_group(df, group1_cols):

    p_values = []
    print(group1_cols)
    for i, row in df.iterrows():
        g1 = row[group1_cols].values.astype(float)
        res = ttest_1samp(g1, popmean=0.5, axis=-1)
        p_values.append(res.pvalue)

    print(f"T-statistic: {res.statistic}, P-value: {res.pvalue}")

    return p_values


def collect_features():

    print(transpose_df.head())

    combined_df = pd.DataFrame()

    features_l = [Tspan8_lEV, Tspan8_sEV, Tspan8_EVdep, Tspan8_F1, Tspan8_F3, V_lEV, V_sEV, V_EVdep, V_F1, V_F3, Tspan8_F1, Tspan8_F3, F1, F3, V_F1, V_F3] 
    for idx, item in enumerate(features_l):
        c_name = item[0]
        print(f"Processing {c_name} ...")
        df_condition = transpose_df[item]
        df_condition = df_condition.replace(0, np.nan)
        df_condition = df_condition[item].replace({',': ''}, regex=True).astype(float)
        mean = df_condition.mean(axis=1)
        df_condition[f"mean_{c_name}"] = mean.tolist()
        p_values = compute_pvalues_one_group(transpose_df, item)
        df_condition["p_value"] = p_values
        path = output_path + "proteomic_data_{}.csv".format(c_name)
        print(path)
        df_condition.to_csv(path, sep="\t", index=None)
        if combined_df is None:
            combined_df = df_condition
        else:
            combined_df = pd.concat([combined_df, df_condition], axis=1)
    combined_path = output_path + "proteomic_data_combined_p_values.csv"
    combined_df.to_csv(combined_path, sep="\t", index=None)


collect_features()
```

```python
import umap
import seaborn as sns
import matplotlib.pyplot as plt

def prepare_data_for_umap():
    features_path = output_path + "proteomic_data_combined_p_values.csv"
    features_mean_df = pd.read_csv(features_path, sep="\t")
    proteomic_features = []
    labels = []
    features_l = [Tspan8_lEV, Tspan8_sEV, Tspan8_EVdep, Tspan8_F1, Tspan8_F3, V_lEV, V_sEV, V_EVdep, V_F1, V_F3, Tspan8_F1, Tspan8_F3, F1, F3, V_F1, V_F3] 
    for idx, item in enumerate(features_l):
        c_name = item[0]
        print(f"Processing {c_name} ...")
        mean_col_name = f"mean_{c_name}"
        print(mean_col_name)
        df_condition = features_mean_df[mean_col_name].replace(np.nan, 0.0)
        print(df_condition)
        proteomic_features.append(df_condition.values)
        labels.append(features_l[idx][0])
    print(proteomic_features[0])
    print(labels[0])
    return proteomic_features, labels

features, labels = prepare_data_for_umap()
```

```python
n_neighbors=20 #10 #5
min_dist=0.8 #0.99 #0.3
metric='correlation'

def plot_umap_features(features, labels):
    embeddings = umap.UMAP().fit_transform(features) # n_neighbors=n_neighbors, min_dist=min_dist, metric='correlation'
    # Create a DataFrame with UMAP components and labels
    data = {"UMAP1": embeddings[:, 0], "UMAP2": embeddings[:, 1], "Label": labels}
    df = pd.DataFrame(data)
    plt.figure(figsize=(8, 6))

    sns.scatterplot(x="UMAP1", y="UMAP2", hue="Label", data=df, palette="viridis", s=50, alpha=1.0)
    plt.title("UMAP Visualization of proteomic Features")
    plt.savefig(output_path + "umap_NeDBIT_features_{}_{}.pdf".format(n_neighbors, min_dist))
    plt.show()

plot_umap_features(features, labels )
```
