```python
import numpy as np
from bioinfokit import analys, visuz
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def compute_log2fc(df, condition_A, condition_B, pseudocount=0.01):
    # Convert percentages to proportions and add pseudocount
    prop_A = (df[condition_A] / 100) + pseudocount
    prop_B = (df[condition_B] / 100) + pseudocount
    
    # Compute log2 fold change
    log2fc = np.log2(prop_B / prop_A)
    
    return log2fc


def plot_MA(df, name):
    
    df["M"] = np.log2((df["value2"] + 1) / (df["value1"] + 1))
    df["A"] = 0.5 * (np.log2(df["value1"] + 1) + np.log2(df["value2"] + 1))
    
    plt.figure(figsize=(8, 6))
    plt.scatter(df["A"], df["M"], alpha=0.5)
    plt.axhline(y=0, color="red", linestyle="--")  # Reference line at M=0
    
    # Annotate points exceeding threshold
    for i, row in df.iterrows():
        plt.annotate(row["Modification_names"], (row["A"], row["M"]), fontsize=9, alpha=0.8)

    plt.xlabel("A (Mean Expression)")
    plt.ylabel("M (Log2 Fold Change)")
    plt.title("MA Plot for {}".format(name))
    plt.grid(True)
    plt.savefig(output_path + "MA_plot_for_{}.png".format(name))
    plt.show()


'''
EV_dep = ["V_dep...Histone19_C3", "V_dep...Histone20_D3", "V_dep...Histone21_E3"]
EV_F1 = ["V_F1...Histone40_H5", "V_F1...Histone41_A6", "V_F1...Histone42_B6"]
EV_F3 = ["V_F3...Histone46_F6",	"V_F3...Histone47_G6", "V_F3...Histone48_H6"]
EV_L = ["V_L...Histone13_E2", "V_L...Histone14_F2", "V_L...Histone15_G2"]
EV_S = ["V_S...Histone16_H2", "V_S...Histone17_A3",	"V_S...Histone18_B3"]

T8_dep = ["T8_dep...Histone28_D4", "T8_dep...Histone29_E4", "T8_dep...Histone30_F4"]
T8_F1 = ["T8_F1...Histone49_A7", "T8_F1...Histone50_B7", "T8_F1...Histone51_C7"]
T8_F3 = ["T8_F3...Histone55_G7", "T8_F3...Histone56_H7", "T8_F3...Histone57_A8"]
T8_L = ["T8_L...Histone22_F3", "T8_L...Histone23_G3", "T8_L...Histone24_H3"]
T8_S = ["T8_S...Histone25_A4", "T8_S...Histone26_B4", "T8_S...Histone27_C4"]

'''

def compute_differential_modifications():
    agg_condition = "mean"
    file_name = output_path + "differential_histone_modification.xlsx"
    V_conditions = [EV_L, EV_S, EV_dep, EV_F1, EV_F3, T8_F1, T8_F3, EV_F1, EV_F3] #[EV_dep, EV_F1_2_3, EV_F4_5, EV_G, EV_L, EV_S]
    T8_conditions = [T8_L, T8_S, T8_dep, T8_F1, T8_F3, T8_S, T8_dep, EV_S, EV_dep] #[T8_dep, T8_F1_2_3, T8_F4_5, T8_G, T8_L, T8_S]

    #T8_con1 = []
    #T8_con2 = []
 
    
    writer = pd.ExcelWriter(file_name, engine="xlsxwriter")
    for idx, v_item in enumerate(V_conditions):
        v_name = v_item[0].split("...")[0]
        df_V_condition = transpose_df[v_item]
        df_V_condition = df_V_condition.replace(0, np.nan)
        mean_v = df_V_condition.mean(axis=1)
        df_V_condition["mean"] = mean_v.tolist()
        
        #print(df_V_condition)
        #print()
        
        t8_item =  T8_conditions[idx]
        t8_name = T8_conditions[idx][0].split("...")[0]
        df_t8_condition = transpose_df[t8_item]
        df_t8_condition = df_t8_condition.replace(0, np.nan)
        mean_t8 = df_t8_condition.mean(axis=1)
        df_t8_condition["mean"] = mean_t8.tolist()
        
        #print(df_t8_condition)
        #print()
        
        df_T8_V = pd.concat([df_t8_condition[agg_condition], df_V_condition[agg_condition]], axis=1)
        df_T8_V.columns = [t8_name, v_name]
        
        df_T8_V["modification_names"] = modification_names
        
        #print(df_T8_V)
        #print()
        
        df_T8_V = df_T8_V[(df_T8_V[t8_name] > 0) & (df_T8_V[v_name] > 0)]
        
        log2_name = "log2FC_{}/{}".format(t8_name, v_name)
        df_T8_V[log2_name] = compute_log2fc(df_T8_V, t8_name, v_name)
        
        df_T8_V = df_T8_V[(df_T8_V[log2_name] < -1) | (df_T8_V[log2_name] >= 1)]
        df_T8_V = df_T8_V.sort_values(by=log2_name, ascending=False)
        
        #print(df_T8_V)
        #print()
        # T8.dep     EV.dep modification_names  log2FC_T8.dep/EV.dep
        df_T8_V_vis = df_T8_V.copy()
        df_T8_V_vis.columns = ["value1", "value2", "Modification_names", "log2FC"]
        
        #visuz.GeneExpression.ma(df=df_T8_V_vis, lfc="log2FC", ct_count="value1", st_count="value2", show=True)
        
        tab_name = "{}_VS_{}".format(t8_name, v_name)
        plot_MA(df_T8_V_vis, tab_name)
        df_T8_V.to_csv(output_path + "log2FC_histone_modifications_{}.csv".format(tab_name), sep="\t", index=None)
        #print(tab_name)
        df_T8_V.to_excel(writer, sheet_name=tab_name, index=False)
        #break
    writer.close()
compute_differential_modifications()
```
