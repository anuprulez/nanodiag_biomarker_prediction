```python
import numpy as np

def compute_log2fc(df, condition_A, condition_B, pseudocount=0.01):
    # Convert percentages to proportions and add pseudocount
    prop_A = (df[condition_A] / 100) + pseudocount
    prop_B = (df[condition_B] / 100) + pseudocount
    
    # Compute log2 fold change
    log2fc = np.log2(prop_B / prop_A)
    
    return log2fc


def compute_differential_modifications():
    agg_condition = "mean"
    file_name = output_path + "differential_histone_modification.xlsx"
    l_conditions = [EV_dep, EV_F1_2_3, EV_F4_5, EV_G, EV_L, EV_S, T8_dep, T8_F1_2_3, T8_F4_5, T8_G, T8_L, T8_S]
    writer = pd.ExcelWriter(file_name, engine="xlsxwriter")
    for item in l_conditions:
        c_name = item[0].split("...")[0]
        df_curr_condition = transpose_df[item]
        mean_control = df_curr_condition.mean(axis=1)
        df_curr_condition["mean"] = mean_control.tolist()
        df_curr_condition_ctrl = pd.concat([df_control[agg_condition], df_curr_condition[agg_condition]], axis=1)
        df_curr_condition_ctrl.columns = ["control", c_name]
        
        df_curr_condition_ctrl["modification_names"] = modification_names
        
        #print(df_curr_condition_ctrl)
        #print()
        
        df_curr_condition_ctrl = df_curr_condition_ctrl[(df_curr_condition_ctrl["control"] > 0) & (df_curr_condition_ctrl[c_name] > 0)]
        
        #print(df_curr_condition_ctrl)
        #print()
        
        log2_name = "log2FC_{}/control".format(c_name)
        df_curr_condition_ctrl[log2_name] = compute_log2fc(df_curr_condition_ctrl, "control", c_name)
        
        #print(df_curr_condition_ctrl)
        #print()
        
        df_curr_condition_ctrl = df_curr_condition_ctrl[(df_curr_condition_ctrl[log2_name] < -1) | (df_curr_condition_ctrl[log2_name] >= 1)]
        df_curr_condition_ctrl = df_curr_condition_ctrl.sort_values(by=log2_name, ascending=False)
        
        #print(df_curr_condition_ctrl)
        #print()
        tab_name = "{}_VS_Ctrl".format(c_name)
        df_conditions_ctrl_ev_dep.to_csv(output_path + "log2FC_histone_modifications_{}.csv".format(tab_name), sep="\t", index=None)
        print(tab_name)
        df_curr_condition_ctrl.to_excel(writer, sheet_name=tab_name, index=False)
        #break
    writer.close()
compute_differential_modifications()
```
